```python
%load_ext autoreload
%autoreload 2

import torch
import numpy as np
import matplotlib.pyplot as plt

import os
import PIL.Image as Image
import cv2 as cv
from tqdm import tqdm
from utils import imshow

import pickle

from pose_utils import get_model, get_poses
```

```python
get_model()
```

```python
IMG_DIR = "data"
OUT_DIR = "pose_out"

poses = {}

for filename in tqdm([s for s in os.listdir(IMG_DIR) if s.endswith(".jpg")]):
	name = os.path.splitext(filename)[0]
	
	img = cv.imread(os.path.join(IMG_DIR, filename))
	pose = get_poses(img)
	
	pose.save(os.path.join(OUT_DIR, f"pose_{name}.png"))
	poses[name] = pose
```

```python
with open("pose_out/pose_preds.pkl", "wb") as f:
	pickle.dump(poses, f)
```
