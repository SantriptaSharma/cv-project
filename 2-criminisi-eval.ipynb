## Observations (from behind the scenes)
- The chosen pairs of parallel lines for each axis highly affect the accuracy of measurements. 
- The choice of reference height does not affect the relative error too greatly, but does affect the direction of the error.
- Hypothesis: The accuracy is inversely proportional (with some degree) to the size of the vanishing points. That is, if the scene has an axis (and corr. || lines) which intersect "quickly" in the image plane, such as the ones given by two parallel lines in the scene that are vastly different in depth, accuracy is high, otherwise accuracy is low.
- Consequently, fiddling with the axes and parallel lines selection can often increase accuracy. However, in some cases, there is no good choice of axis.
- This makes this method difficult to apply to arbitrary images, not to mention time consuming.
- Another limitation is that we can only measure certain kinds of distances. Here, I've implemented measuring between parallel planes, but although Criminisi's methods allow for more modes of measurement, measuring an arbitrary line off/unaligned/orthogonal to the reference plane is not possible.
- To ameliorate some of these limitations, we move to using an MMDE framework for this task.

We can do a quick test of our hypothesis:

```python
annot_files = [s for s in os.listdir(BASE_DIR) if s.startswith("kartripta") and "-annot.npz" in s]

names = [s.split("-annot")[0] for s in annot_files]

names
```

```python
mean_errors = []
mean_rel_errors = []

mean_vps = []
max_vps = []

norm_vps_x = []
norm_vps_y = []
norm_vps_z = []

for name in names:
	origin, axes, parallel_lines = load_axes(name)
	lines, measurements = load_measurements(name)

	vps = compute_vanishing_points(parallel_lines)

	normed_vps = np.linalg.norm(vps, axis=1)

	mean_vps.append(np.mean(normed_vps))
	max_vps.append(np.max(normed_vps))

	norm_vps_x.append(normed_vps[0])
	norm_vps_y.append(normed_vps[1])
	norm_vps_z.append(normed_vps[2])

	P = get_proj_mat(vps)
	alpha = establish_scale(lines[0], P, measurements[0])

	N = len(lines)

	abs_sum = 0
	rel_sum = 0
	for i in range(1, N):
		height = compute_height(lines[i], P, alpha)

		abs_err = np.abs(height - measurements[i])
		rel_err = abs_err / measurements[i]
		
		abs_sum += abs_err
		rel_sum += rel_err

	mean_errors.append(abs_sum / (N - 1))
	mean_rel_errors.append(rel_sum / (N - 1))
```

```python
mean_rel_errors[-1]
```

```python
index = list(range(len(mean_errors)))

fig, ax1 = plt.subplots()

ax1.set_label("image index")
ax1.set_ylabel("mean error (%)")
ax1.plot(index, mean_rel_errors, color="tab:red", label = "mean errors (%)")
ax1.tick_params(axis="y", color="tab:red")

ax2 = ax1.twinx()
ax2.set_ylabel("vanishing point magnitude (z)")
ax2.plot(index, norm_vps_z, color="tab:green", label = "z vanishing point magnitude")
ax2.tick_params(axis="y", color="tab:green")
# ax2.set_yscale("log")

fig.legend()
```

Which reveals that the hypothesis doesn't seem to hold up. An alternate hypothesis is that what is more relevant is the world's relative rotation wrt to the camera plane, with a high angle b/w the plane and the world frame giving better results. 

This makes intuitive sense, since in the case where the image plane is completely aligned with the induced world frame (looking directly at the front of a building, for example), we would have a degenerate case and wouldn't be able to identify any vanishing points but one.

It stands to reason (by some messed up, nonrigorous, etc. application of a continuity argument), that as we increase this angle away from 0, the overall error should decrease.

We can test this since we have P up to some unknown scale, which does not affect the rotation.
