## Axes Annotations

```python
%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
import os

from criminisi import select_parallel_lines_and_axis, draw_axis
from utils import imshow, select_lines, draw_lines

BASE_IN = "data"
BASE_OUT = "criminisi_data"
```

```python
name = "torch_2"
im = cv.imread(os.path.join(BASE_IN, name + ".jpg"))
```

```python
%matplotlib Qt
origin, axes, lines = select_parallel_lines_and_axis(im)
```

```python
%matplotlib inline
axis_im = draw_axis(im, origin, axes)

imshow(axis_im)
```

```python
inp = input("save (y/n):")

if inp[0] == "y":
	np.savez(os.path.join(BASE_OUT, name + "-annot.npz"), origin=origin, axes=axes, lines=lines)
	print("saved")
```
