## Axes Annotations

```python
%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
import os

from criminisi import select_parallel_lines_and_axis, draw_axes, save_axes, save_measurements
from utils import imshow, select_lines, draw_lines

BASE_IN = "data"
BASE_OUT = "criminisi_data"
```

```python
name = "kartripta12"
im = cv.imread(os.path.join(BASE_IN, name + ".jpg"))
```

```python
%matplotlib Qt
origin, axes, lines = select_parallel_lines_and_axis(im)
```

```python
origin
```

```python
%matplotlib inline
axis_im = draw_axes(im, origin, axes)

imshow(axis_im)
```

```python
inp = input("save (y/n):")

if inp[0] == "y":
	save_axes(name, origin, axes, lines)
	print("saved")
```
